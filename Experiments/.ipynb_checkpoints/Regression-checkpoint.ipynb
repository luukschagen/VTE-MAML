# Regression Plots

```python
def plot_axis(plt):
    plt.axis('on')
    x_range = torch.arange(-5,5,0.1).view(-1,1)
    x_s, x_q, y_s, y_q = testdata[0]
    preds = []
    for i in range(100):
        args = vte.finetune(x_s, y_s)
        with torch.no_grad():
            preds.append(vte.predict(x_range, *args))

    # x_q, ind = x_q.sort(dim=0)
    # preds = [pred[ind].view(-1,1) for pred in preds]
    args = mmaml.finetune(x_s, y_s)
    altpred = mmaml.predict(x_range, *args)

    args = maml.finetune(x_s, y_s)
    mamlpred = maml.predict(x_range, args)

    for pred in preds:
        plt.plot(x_range, pred.detach(), c='grey', alpha=0.1)
    plt.scatter(x_q, y_q.view(-1,1), s=8)
    plt.plot(x_range, altpred.detach(), c='red')
    plt.plot(x_range, mamlpred.detach(), c='blue')
    plt.scatter(x_s, y_s, s=100, marker='^', c='purple', zorder=100)
```

```python
fig, axes = plt.subplots(10, 5, figsize=(15,21))

for i in range(10):
    for j in range(5):
        plot_axis(axes[i,j])

plt.show()
```

```python
fig.savefig("regression_mosaic_2.pdf", format='pdf', bbox_inches='tight')
```

# Embedding space

```python
testdata = datasets.MultiModal(100, 5, 100, 3)

embeddings = None
for i in range(1000):
    with torch.no_grad():
        x_s, __, y_s, __ = testdata[0]
        embed = vte.modulation_model.transform_encoding(vte.modulation_model.encode(x_s, y_s))[0]
        if embeddings is None:
            embeddings = embed.view(1,-1)
        else:
            embeddings = torch.cat([embeddings, embed], dim=0)
```

```python
tsne = TSNE()
```

```python
tsne = tsne.fit_transform(embeddings)
```

```python
ls = []
for i in range(333):
    ls += [1,2,3]
ls.append(1)
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(tsne[:,0], tsne[:,1], c=ls, s=10, alpha=1)
legend = ax.legend(scatter.legend_elements()[0], ["sinusoid", "linear", "quadratic"])
ax.add_artist(legend)
plt.savefig("tsne.pdf", format="pdf")
```

# Convergence analysis

```python
def test_convergence(metalearner, n_steps, testdata, samples):
    scores = []
    for i in range(n_steps):
        scores.append(evaluation.score(metalearner, testdata, samples, i, metric='mse'))
    return scores
```

```python
mmaml_values = test_convergence(mmaml, 10, testdata, 1)
```

```python
maml_valus = test_convergence(maml, 10, testdata, 1)
```

```python
vte_values = test_convergence(vte, 10, testdata, 5)
```

```python
plt.plot(list(range(10)), maml_valus, marker="x")
plt.plot(list(range(10)), mmaml_values, marker="^")
plt.plot(list(range(10)), vte_values, marker="*")
plt.legend(['MAML', 'Multimodal MAML', 'VTE-MAML'])
plt.xlabel("Number of inner gradient steps", fontsize=16)
plt.ylabel("Mean Squared Error", fontsize=16)
plt.savefig("convergence.pdf", format="pdf")
```
